```python
#   XOR.py-A very simple neural network to do exclusive or.
import numpy as np
 
epochs = 100       # Number of iterations
inputLayerSize, hiddenLayerSize, outputLayerSize = 64, 25, 1
#  Xt = X_q.popleft()
# X = np.array([[0,0], [0,1], [1,0], [1,1]])
# y = np.array([ [0],   [1],   [1],   [0]])
 
def sigmoid (x): return 1/(1 + np.exp(-x))      # activation function
def sigmoid_(x): return x * (1 - x)             # derivative of sigmoid
                                                # weights on layer inputs
Wh = np.random.uniform(size=(inputLayerSize, hiddenLayerSize))
Wz = np.random.uniform(size=(hiddenLayerSize,outputLayerSize))
 
error = []
ydash= []; whidden = []; wop = []
for i in range(epochs):
    for p,q in zip(X1,y1):
#         print p
#         print "inner", i
        H = sigmoid(np.dot([p], Wh))                  # hidden layer results
        Z = sigmoid(np.dot(H, Wz))                  # output layer results
#         import pdb;pdb.set_trace()
        E = np.square(q - Z)                                   # how much we missed (error)
        error.append(E)
        dZ = E * sigmoid_(Z)                        # delta Z
        dH = dZ.dot(Wz.T) * sigmoid_(H)             # delta H
    #     import pdb;pdb.set_trace()
        Wz +=  H.T.dot(dZ)                          # update output layer weights
        Wh +=  np.array([p]).T.dot(dH)                          # update hidden layer weights
        ydash.append(Z[0][0])
print(Z)        
```

```python

def infer(input):
    H = sigmoid(np.dot(input, Wh))
    Z = sigmoid(np.dot(H, Wz))
    return Z
# np.square([1,2,3])
```
